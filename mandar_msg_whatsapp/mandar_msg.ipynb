```python
keyboard = Controller()

def enterChat(chatNameSearch):
    search = nav.find_element(By.CSS_SELECTOR, 'button[data-testid=icon-search-morph]')
    search.click()
    
    text = nav.find_element(By.CSS_SELECTOR, '[data-testid=chat-list-search]')
    text.send_keys(chatNameSearch)

    time.sleep(2)

    text.send_keys(Keys.ENTER)
    
    cancelarPesquisa = nav.find_element(By.CSS_SELECTOR, '[aria-label="Cancelar pesquisa"]')
    cancelarPesquisa.click()

    text.send_keys(Keys.ESCAPE)
    
    time.sleep(1)



def enterNthNumber(index):

    enterChat("Só eu")

    messages = nav.find_elements(By.CSS_SELECTOR, '[data-testid=msg-container]:last-child')
    ultimaMsg = None
    for msg in messages:
        ultimaMsg = msg

    linkTags = ultimaMsg.find_elements(By.CSS_SELECTOR, "a")
    links = []
    for linkTag in linkTags:
        links.append(linkTag)

    links[index].click()

    # time.sleep(10)
    conversarCom = WebDriverWait(nav, timeout=10).until(lambda d: d.find_element(By.CSS_SELECTOR, '[data-testid=mi-message-on-whatsapp]'))
    print("achou")

    # conversarCom = nav.find_element(By.CSS_SELECTOR, '[data-testid=mi-message-on-whatsapp]')
    conversarCom.click()

def sendText(text):
    chatBox = nav.find_element(By.CSS_SELECTOR, '[data-testid=conversation-compose-box-input]')
    chatBox.send_keys(text)
    chatBox.send_keys(Keys.ENTER)


def sendImage(text="", imgWinV=None, textWinV=None):
    chatBox = nav.find_element(By.CSS_SELECTOR, '[data-testid=conversation-compose-box-input]')

    if imgWinV is None:
        chatBox.send_keys(Keys.CONTROL, "v")
    else: 
        chatBox.click()
        winPaste(imgWinV)

    legendaBox = WebDriverWait(nav, timeout=10).until(lambda d: d.find_element(By.CSS_SELECTOR, '[data-testid=media-caption-input-container]'))
    if textWinV is None:
        legendaBox.send_keys(text)
    else:
        winPaste(textWinV)
        time.sleep(1)

    legendaBox.send_keys(Keys.ENTER)

def winPaste(num):
    

    keyboard.press(Key.cmd)
    keyboard.tap('v')
    keyboard.release(Key.cmd)

    time.sleep(0.5)
    for n in range(num):
        keyboard.tap(Key.down)

    keyboard.tap(Key.enter)
    time.sleep(0.5)

messagesBefore = []
messagesAfter = []
messagesBefore.append("Oi, tudo bem? Você é bixo de elétrica 023 certo?")
messagesBefore.append("Primeiramente parabéns!")
messagesBefore.append("Meu nome é Artur, sou de uma república, nós estamos com vagas abertas")
messagesBefore.append("Vou encaminhar uma mensagem de divulgação da república")
messagesAfter.append("Caso queira fazer alguma pergunta ou marcar um dia para visita fique a vontade")
```
